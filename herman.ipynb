```python
class Node:
    TITLE_IDX=0
    ARTIST_IDX=1
    GENRE_IDX=2
    YEAR_IDX=3
    def __init__(self, initial_data):
        self.data = initial_data
        data_split = initial_data.split("_")
        self.title = data_split[self.TITLE_IDX]
        self.artist = data_split[self.ARTIST_IDX]
        self.year = data_split[self.YEAR_IDX]
        self.next = None
    

#class definitions sourced from zybooks and geeks from geeks, then altered
class LinkedList:
    def __init__(self):
        self.head = None
        self.tail = None

    def append(self, new_node):
        if self.head == None:
            self.head = new_node
            self.tail = new_node
        else:
            self.tail.next = new_node
            self.tail = new_node

    def prepend(self, new_node):
        if self.head == None:
            self.head = new_node
            self.tail = new_node
        else:
            new_node.next = self.head
            self.head = new_node

    def insert_after(self, current_node, new_node):
        if self.head == None:
            self.head = new_node
            self.tail = new_node
        elif current_node is self.tail:
            self.tail.next = new_node
            self.tail = new_node
        else:
            new_node.next = current_node.next
            current_node.next = new_node
   
    def remove_after(self, current_node):
        print("Current node=",current_node)
        
        print("current==None", current_node==None)
        print("self.head != None", self.head != None)
        if (current_node == None) and (self.head != None):
            succeeding_node = self.head.next
            self.head = succeeding_node  
            if succeeding_node == None: 
                self.tail = None
        
        elif current_node.next != None:
            print("current_node.next != None", current_node.next != None)
            print("current_node.next",current_node.next )
            succeeding_node = current_node.next.next
            current_node.next = succeeding_node
            if succeeding_node == None: 
                self.tail = current_node

    def printList(self):
      # Output list
        print('LINKED LIST:', end=' ')
        node = self.head
        while node != None:
          print(node.data, end='->')
          node = node.next
        print()

    def search(self, title):
 
        # Initialize current to head
        current = self.head
 
        # loop till current not equal to None
        while current != None:
            if current.title == title:
                return True # data found
             
            current = current.next
         
    
        return False # Data Not found

#Default class definitions end here       


#Reads csv into singly-linked list
def main():
    topPlays = LinkedList()
    raw_file = ''
    
    with open('topSpotify2010-2019.txt','r', encoding = 'utf-16') as file:
        raw_file = file.read()

    i=0
    for entry in raw_file.split("\n"):
        if i > 0 and entry:
            node = Node(entry)
            
            topPlays.append(node)
        i+=1
    
    node = topPlays.head
    print(node.artist)
    print(node.next.data)
    
#     print(topPlays)
#     while topPlays !=' ':
#         spotified = spotified.rstrip('\n')
#         node = Node(spotified)
#         topPlays.append(node)
#         spotified = infile.readline()
#     infile.close()
 

    print('List after adding nodes:', end=' ')
    node = topPlays.head
    while node != None:
        print(node.artist)
        node = node.next

main()

```
